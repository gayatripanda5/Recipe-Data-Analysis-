```python
#Scraping Gobi Manchurian recipe
from bs4 import BeautifulSoup
import requests
# Request to website and download HTML contents
url='https://www.sanjeevkapoor.com/recipe/Gobi-Manchurian.html'
req=requests.get(url)
content=req.text
soup = BeautifulSoup(content, 'html.parser')
ingredients = soup.find_all('div', class_ = 'ingredientlist')
method=soup.find_all('div', class_ = 'stepdetail')

################### Scraping Ingredient List########################

ing=ingredients[0]
g=ing.ul.find_all('span')
f=open("Recipe of Gobi Manchurian.txt", 'w+')
str1="-----------------Recipe of Gobhi Manchurian------------------"
f.write(str1)
f.write("\nIngredient List\n")
for i in range(len(g)):
    f.write( "%s" %(g[i].text))
    f.write("\n")
f.close()
 
####################Cooking Instructions steps######################

steps=[]
for step in method:
    steps.append(step.text)
#print(steps)
############################## Saving in a file ###########

f= open("Recipe of Gobi Manchurian.txt",'a')
f.write("\nCooking Instructions\n")
for k in range(len(steps)):
    f.write("%s. %s" %(k+1,steps[k]))
    f.write("\n")
f.close()
```

<font size='3'><b>Q3.(a) Plot the frequency-rank distribution for all the recipes and interpret.</b></font>

```python
#ingredients
df_all_ing=pd.DataFrame(data=ingredients,columns=['Ingredient_name'])
df_all_ing.head()
series_ing=df_all_ing['Ingredient_name'].value_counts()
df_ing=series_ing.rename_axis('Ingredient_name').reset_index(name='frequency')
df_ing['Rank'] = np.arange(1,len(df_ing)+1)
len(df_ing)
df_ing.head()
```

```python
df_ing.frequency[2000]
```

```python
ax1=df_ing.frequency.plot.line()
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.title("Ingredient Frequency Rank Distribution")
plt.gcf().set_size_inches(10, 10)
```

```python
#loglog plot of Freq-Rank Distribution

x=df_ing.Rank
y=df_ing.frequency
#plt.loglog(x,y, basex=np.e, basey=np.e)

fig, ax = plt.subplots()

ax.loglog(x,y, basex=np.e, basey=np.e,marker=".")
def ticks(y, pos):
    return r'$e^{:.0f}$'.format(np.log(y))

ax.xaxis.set_major_formatter(mtick.FuncFormatter(ticks))
ax.yaxis.set_major_formatter(mtick.FuncFormatter(ticks))

plt.gcf().set_size_inches(7, 7)
plt.xlabel('Rank')
plt.ylabel('Normalized Frequency')
```

<font size='3'><b>INTERPRETATION:</b> </font><font size='3'>The ingredient frequency-rank distribution has 6714 ingredients ranked(x-axis) on the basis of their frequency(y-axis) in all the cuisines. This "L-shaped" curve follows "L-curve" of Power Law, with a tall-head and a long-tail. It can only be monotonously decreasing function. It can be observed that the difference between the most frequent and least frequent ingredient is quite large(~18K times). Even the difference among the more frequent ingredients is also large indicated by this huge dip in frequency till 1000th rank, after that frequency of almost all the ingredients seems similar represented by a nearly straight line indicating that certain ingredients are used more frequently used.</font>
