```python
# import library
from bs4 import BeautifulSoup
import requests
# Request to website and download HTML contents
url='https://www.sanjeevkapoor.com/recipe/Aloo-Bhujiya.html'
req=requests.get(url)
content=req.text
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup=BeautifulSoup(content)
```

```python
ingredients = soup.find_all('div', class_ = 'ingredientlist')
method=soup.find_all('div', class_ = 'stepdetail')
```

```python
ing=ingredients[0]
```

```python
#ingredient list

g=ing.ul.find_all('span')
f=open("Recipe of Aloo-bhujiya.txt", 'w+')
str1="-----------------Recipe of Aloo-bhujiya------------------"
f.write(str1)
f.write("\nIngredient List\n")
#f.write("\n")
for i in range(len(g)):
    f.write("%s. %s" %(i+1,g[i].text))
    f.write("\n")
f.close()
    #ing_list.append(i.text)
#print(ing_list)
```

```python
#method steps
steps=[]
for step in method:
    steps.append(step.text)
print(steps)
```

```python
ing=ingredients[0]
```

```python
f=open("Recipe of Aloo-bhujiya.txt", 'a')
f.write("\nCooking Instructions\n")
for k in range(len(steps)):
    f.write("%s. %s" %(k+1,steps[k]))
    f.write("\n")
f.close()
```
